```python
vocabs = {}
i = 0

for word in wsj.model.vocab:
    if word in google.model.vocab and word in wiki.model.vocab:
        if i % 100 is 0:
            logger.info('%d: %s' % (i, word))
            pickle.dump(vocabs, open('vocabs_reference_%d.pkl' % i, 'wb'))
        wsj_set = set([a[0] for a in wsj.get_similar_words(word)])    
        google_set = set([a[0] for a in google.get_similar_words(word)])
        wiki_set = set([a[0] for a in wiki.get_similar_words(word)])
    
        google_count = len(wsj_set & google_set)
        wiki_count = len(wsj_set & wiki_set)
        
        vocabs[word] = {'google_count': google_count, 'wiki_count': wiki_count}
    i += 1

pickle.dump(vocabs, open('vocabs_reference.pkl', 'wb'))
```

```python
vocabs_loaded = pickle.load(open('vocabs_reference.pkl', 'r'))

wsj_substitute_key_inv = {v: k for k, v in wsj_substitute_key.items()}

for word, counts in vocabs_loaded.iteritems():
    vocabs_loaded[word]['is_codeword'] = word in wsj_substitute_key_inv
```

## Using Google News as a Reference Corpus

We compare the **substituted WSJ corpus** against **Google News** reference corpus and find the following result.

```python
import numpy

google_counts_codewords = [counts['google_count'] for word, counts in vocabs_loaded.iteritems() if counts['is_codeword']]
google_counts_usual = [counts['google_count'] for word, counts in vocabs_loaded.iteritems() if not counts['is_codeword']]

f, axes = plt.subplots(2, sharex=True)
f.suptitle('Google')
axes[0].set_title('Non-Codewords')
axes[1].set_title('Codewords')
axes[0].set_ylabel('Word Count')
axes[1].set_ylabel('Word Count')
bins = numpy.linspace(0, 200, 100)
axes[0].hist(google_counts_usual, bins)
axes[1].hist(google_counts_codewords, bins)
plt.show()
```

## Using WSJ as a Reference Corpus

We compare the **substituted WSJ corpus** against **the original WSJ corpus** reference corpus and find the following result. This is to remind readers what an ideal distribution should look like.

```python
import lib.model_loaders.codeword_model_loader as codeword_model_loader

wsj_substituted_model = codeword_model_loader.CodewordModel('./models/wsj_substituted_w2v.mdl')
wsj_original_model = codeword_model_loader.CodewordModel('./models/wsj_original_w2v.mdl')

vocabs_wsj = {}

for word in wsj_substituted_model.model.vocab:
    sub_set = set([a[0] for a in wsj_substituted_model.get_similar_words(word)])
    org_set = set([a[0] for a in wsj_original_model.get_similar_words(word)])
    vocabs_wsj[word] = {'wsj_count': len(sub_set & org_set)}

vocabs_wsj
```

```python
def summary_google():
    temp_substitute_key = pickle.load(open('substitute_keys/wsj_substitute_key.pkl'))
    vocabs_loaded = pickle.load(open('wsj_vocabs_reference/vocabs_reference.pkl', 'r'))

    temp_substitute_key_inv = {v: k for k, v in temp_substitute_key.items()}
    
    for word, counts in vocabs_loaded.iteritems():
      vocabs_loaded[word]['is_codeword'] = word in temp_substitute_key_inv
    
    words = [{'word': word, 'google_count': vocabs_loaded[word]['google_count'], 'wiki_count': vocabs_loaded[word]['wiki_count'], 'is_codeword': vocabs_loaded[word]['is_codeword']} for word in vocabs_loaded.iterkeys()]
    
    words = sorted(words, key=lambda k: k['google_count'])
    
    total_codeword_count = len(temp_substitute_key)
    
    true_positive = 0 # selected as codeword and is codeword
    true_negative = len(words) - len(temp_substitute_key) # not selected as codeword and is not codeword
    false_positive = 0 # selected as codeword and isn't codeword
    false_negative = len(temp_substitute_key) # not selected as codeword and is codeword
    
    precision_recall_f1s = []
    
    for word_count in range(0, len(words)):
        word_tuple = words[word_count]
        if word_tuple['is_codeword']:
            true_positive += 1
            false_negative -= 1
        else:
            false_positive += 1
            true_negative -= 1
        precision = float(true_positive) / float(true_positive + false_positive)
        recall = float(true_positive) / float(true_positive + false_negative)
        try:
            f1 = 2 * precision * recall / (precision + recall)
        except ZeroDivisionError:
            f1 = 0
        
        precision_recall_f1s.append({'precision': precision, 'recall': recall, 'f1': f1, 'x': word_tuple['google_count']})
    
    return precision_recall_f1s

def summary_wiki():
    temp_substitute_key = pickle.load(open('substitute_keys/wsj_substitute_key.pkl'))
    vocabs_loaded = pickle.load(open('wsj_vocabs_reference/vocabs_reference.pkl', 'r'))

    temp_substitute_key_inv = {v: k for k, v in temp_substitute_key.items()}
    
    for word, counts in vocabs_loaded.iteritems():
      vocabs_loaded[word]['is_codeword'] = word in temp_substitute_key_inv
    
    words = [{'word': word, 'google_count': vocabs_loaded[word]['google_count'], 'wiki_count': vocabs_loaded[word]['wiki_count'], 'is_codeword': vocabs_loaded[word]['is_codeword']} for word in vocabs_loaded.iterkeys()]
    
    words = sorted(words, key=lambda k: k['wiki_count'])
    
    total_codeword_count = len(temp_substitute_key)
    
    true_positive = 0 # selected as codeword and is codeword
    true_negative = len(words) - len(temp_substitute_key) # not selected as codeword and is not codeword
    false_positive = 0 # selected as codeword and isn't codeword
    false_negative = len(temp_substitute_key) # not selected as codeword and is codeword
    
    precision_recall_f1s = []
    
    for word_count in range(0, len(words)):
        word_tuple = words[word_count]
        if word_tuple['is_codeword']:
            true_positive += 1
            false_negative -= 1
        else:
            false_positive += 1
            true_negative -= 1
        precision = float(true_positive) / float(true_positive + false_positive)
        recall = float(true_positive) / float(true_positive + false_negative)
        try:
            f1 = 2 * precision * recall / (precision + recall)
        except ZeroDivisionError:
            f1 = 0
        
        precision_recall_f1s.append({'precision': precision, 'recall': recall, 'f1': f1, 'x': word_tuple['wiki_count']})
    
    return precision_recall_f1s
```
